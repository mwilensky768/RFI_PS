```python
from RFI_PS import TophatSource, Instrument
import numpy as np
import matplotlib.pyplot as plt
```

# TV6 Source from paper

```python
def plot_power(k_z, power):
    plt.plot(k_z, power)
    plt.xscale("log")
    plt.yscale("log")
    plt.xlabel(r"$k_\parallel$ $h$ Mpc$^{-1}$")
    plt.ylabel(r"Power (mK$^2$ $h^3$ Mpc$^{-3}$)")
    
```

```python
freqs = np.arange(168.555e6, 187.275e6 + 80e3, 80e3)
beam_width = np.pi / 180 * 7
inst = Instrument(freqs, beam_width)
freq_center = 177.5e6
width = 7e6
flux_total = width
TV6 = TophatSource(flux_total, freq_center, width)

kzs, power = TV6.calc_power(inst)

plot_power(kzs, power)
```

# TV7 source using full MWA band, and much larger beam width

Note this is an extremely wide beam because it is assuming a tophat beam, which has a much larger volume for a given opening angle compared something with the same e.g. FWHM.

```python
starting_freq = 1.28 * 131e6 + 40e3
ending_freq = starting_freq + 384 * 80e3
print(starting_freq, ending_freq)

freqs = np.arange(starting_freq, ending_freq, 80e3)
beam_width = 30 * np.pi / 180. # 30 degree opening angle

inst = Instrument(freqs, beam_width)

# 1 Jy Australian TV7
freq_center = 184.5e6
width = 7e6
flux_total = width

TV7 = TophatSource(flux_total, freq_center, width)
kzs, power = TV7.calc_power(inst)

plot_power(kzs, power)
```
